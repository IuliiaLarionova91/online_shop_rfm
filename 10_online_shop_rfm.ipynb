##  Описание проекта
Интернет-магазин товаров для дома «Пока все ещё тут» нужно сделать расслку товаров заинтересованным покупателям. Для этого нужно понять какие категории покупателей существует. Какой группе покупателей нужено, что-то раccылать, акоторой не нужно. Также нужно определиться, какой товар нужно отправлять каждой определенной группе и в какой сезон.

## Задачи:
* Сегментировать покупателей по профилю потребления
* Провести исследовательский анализ данных
* Сегментировать покупателей на основе истории их покупок
* Проанализировать их покупки и
сегментировать товары на основе принадлежности к той или иной товарной
категории.
* Сформулировать и проверить статистические гипотезы


## Описание данных:
Датасет описывает транзакции интернет-магазина товаров для дома и быта «Пока
все ещё тут».

Колонки:
* date — дата заказа;
* customer_id — идентификатор покупателя;
* order_id — идентификатор заказа;
* product — наименование товара;

```python
def rfm_quantiles(df, column, n_quantiles=3):
    labels = list(range(1,n_quantiles + 1))
    while len(labels)>0:
        try:
            #print(pd.qcut(df[column], n_quantiles, duplicates = 'drop'))
            return pd.qcut(df[column], n_quantiles, labels, duplicates = 'drop').astype('int')
        except:
            labels.pop()
            n_quantiles -= 1
    raise
```

```python
RFM_Segment = RFM_table.copy()

RFM_Segment['R_Quartile'] = rfm_quantiles(RFM_Segment, 'recency')
RFM_Segment['F_Quartile'] = rfm_quantiles(RFM_Segment, 'frequency')
RFM_Segment['M_Quartile'] = rfm_quantiles(RFM_Segment, 'monetary_value')
RFM_Segment['RFMClass'] = RFM_Segment.R_Quartile.map(str) \
                            + RFM_Segment.F_Quartile.map(str) \
                            + RFM_Segment.M_Quartile.map(str)
```

##  Проверим с помощью гипотез корретность сегментации клиентов.
Проверить гипотезу 1.

 - Альтернативная гипотеза: У групп количество дней с последнего заказа отличается.

 - Альтернативная гипотеза: У групп количество чеков отличается.

Проверить гипотезу 3.

 - Нулевая гипотеза: У групп суммарная выручка с покупателя не отличается.
 - Альтернативная гипотеза: У групп суммарная выручка с покупателя отличается.

```python
def significance_test(df, combinations, column):
    result = {'group':[],
             'pvalue': []}
    n_pairs = 0
    for group1 in range(len(combinations)):
        for group2 in range(group1+1, len(combinations)):
            n_pairs += 1
            #print(group1, group2)
            result['group'].append(combinations[group1] + ' ' + combinations[group2])
            try:
                result['pvalue'].append(mannwhitneyu(df[df['RFMClass'] == combinations[group1]][column], df[df['RFMClass'] == combinations[group2]][column]).pvalue)
            except:
                result['pvalue'].append(1)
    
    return pd.DataFrame(result), n_pairs

def approve(row, combinations):
    alpha = 0.05 / combinations
    if row['recency'] < alpha or row['frequency'] < alpha or row['monetary_value'] < alpha:
        return 'H0 отвергается'
    return 'H0 не отвергается' 
```

```python
def test_RFM(df, classes):
    combinations = df[classes].unique()

    recency_test, n_pairs = significance_test(df, combinations, 'recency')
    frequency_test, _ = significance_test(df, combinations, 'frequency')
    monetary_value_test, _ = significance_test(df, combinations, 'monetary_value')

    pvalues = recency_test.merge(frequency_test, on='group').merge(monetary_value_test, on='group')
    pvalues.columns = ['group','recency','frequency','monetary_value']

    pvalues['verdict'] = pvalues.apply(approve, axis = 1, combinations = n_pairs) 

    return pvalues

pvalues = test_RFM(RFM_Segment, 'RFMClass')
pvalues
```

```python
def draw_categories(df, n_top, value_column, ylabel):
    
    # посчитаем среднюю метрику value_column для каждого класса + даты + категории
    segment_month_categ = df.groupby(['RFMClass', 'date_group', 'product_category'])[value_column].mean().to_frame().reset_index().sort_values('date_group')
    segment_month_categ['date_group'] = segment_month_categ['date_group'].dt.to_period('M').astype('str')
    
    # построим графики
    for column in segment_month_categ['RFMClass'].unique():
        
        sub_df = segment_month_categ[segment_month_categ['RFMClass'] == column].sort_values('date_group').drop('RFMClass',axis =1)

        g = sns.catplot(x="date_group", y=value_column, hue='product_category', data=sub_df, kind='point', height=6, aspect=2)
        g.fig.suptitle('Группа ' + column)
        g.set(xlabel='годы', ylabel=ylabel)
        g.set_xticklabels(rotation=30)
            
    plt.show()
    
```

```python
def draw_top_categories(df, n_top, value_column,ylabel):
    
    # посчитаем среднюю метрику value_column для каждого класса + даты + категории
    segment_month_categ = df.groupby(['RFMClass', 'date_group', 'product_category'])[value_column].mean().to_frame().reset_index().sort_values(value_column, ascending = False)
    segment_month_categ['month_product_category'] = segment_month_categ['date_group'].dt.to_period('M').astype('str') + ' ' + segment_month_categ['product_category']
    
    # возьмем топ n_top комбинаций даты + категории для каждой группы покупателей
    segment_top = segment_month_categ.groupby(['RFMClass']).apply(lambda x : x[['month_product_category',value_column]].sort_values(by = value_column, ascending = False).head(n_top).reset_index(drop=True)).reset_index().drop('level_1', axis = 1)
    segment_top['month'] = segment_top['month_product_category'].apply(lambda x : pd.to_datetime(x.split()[0], format = "%Y-%m-%d"))
    segment_top['product_category'] = segment_top['month_product_category'].apply(lambda x :x.split()[1])

    # нарисуем графики
    for column in segment_top['RFMClass'].unique():
        
        g = sns.catplot(x = 'month_product_category', y = value_column, data = segment_top[segment_top['RFMClass'] == column], kind = 'bar', height=3, aspect=5)
        g.fig.suptitle('Группа ' + column)
        g.set(xlabel='товары и месяцы', ylabel=ylabel)
        g.set_xticklabels(rotation=30)
        
    plt.show()
    
```

```python
def get_most_valuable_categories(df, n_top, value_column):
    segment_month_categ = data.groupby(['RFMClass', 'date_group', 'product_category'])[value_column].mean().to_frame().reset_index().sort_values(value_column, ascending = False)
    segment_month_categ['month_product_category'] = segment_month_categ['date_group'].dt.to_period('M').astype('str') + '+' + segment_month_categ['product_category']

    # возьмем топ n_top комбинаций даты + категории для каждой группы покупателей
    segment_top = segment_month_categ.groupby(['RFMClass']).apply(lambda x : x[['month_product_category',value_column]].sort_values(by = value_column, ascending = False).head(n_top).reset_index(drop=True)).reset_index().drop('level_1', axis = 1)
    segment_top['month'] = segment_top['month_product_category'].apply(lambda x : pd.to_datetime(x.split('+')[0], format = "%Y-%m-%d")).dt.month
    segment_top['product_category'] = segment_top['month_product_category'].apply(lambda x :x.split(sep='+')[1])
    
    return segment_top.sort_values(value_column, ascending = False).drop('month_product_category', axis = 1).reset_index(drop=True)
```
